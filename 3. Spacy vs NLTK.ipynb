# NLTK

NLTK stands for Natural Language Toolkit. It is a Python library used for working with human language data (text). NLTK provides easy-to-use interfaces to over 50 corpora and lexical resources, along with a wide range of text-processing libraries for classification, tokenization, stemming, tagging, parsing, and more.

```python
import nltk

# Download tokenizer data
# nltk.download() 

from nltk.tokenize import sent_tokenize, word_tokenize

# Process a text

Text = "Dr. Strange loves pav bhaji of mumbai. Hulk loves chat of delhi"

# Sentence tokenize

Sentence = sent_tokenize(Text)
print('\n Sentence :',Sentence)

# Word tokenize

Word = word_tokenize(Text)
print('\n Word :',Word)
```

# SPACY

spaCy is a fast, modern, and industrial-strength Natural Language Processing (NLP) library for Python. It's designed specifically for production use cases, with a focus on performance, efficiency, and ease of use.

```python
!pip install spacy
```

```python
import spacy

# Load English language model

NLP = spacy.load("en_core_web_sm")

# Process a text

Doc = NLP("Dr. Strange loves pav bhaji of mumbai. Hulk loves chat of delhi")

# To print each sentence from the document

for sentence in Doc.sents:
    print('sentence :', sentence)

# To print each word from the sentence

for sentence in Doc.sents:
    for word in sentence:
        print('\nWord :',word)
        
```
