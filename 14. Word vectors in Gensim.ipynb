# Word vectors in Gensim

Word vectors in Gensim are high-dimensional representations of words generated using models like Word2Vec, FastText, or GloVe. These vectors capture semantic meaning so that similar words have similar vectors.

📦 **Key Idea:**

Word vectors are trained by predicting a word’s context or vice versa using shallow neural networks. Words with similar contexts will have closer vector representations.

```python
# Install the Trained Model

import gensim.downloader as api

wv = api.load("word2vec-google-news-300")  # This is a huge model (~1.6 gb)
```

```python
#  Similarity between two words

S1 = wv.similarity(w1 = "Good", w2 = "Best")
print("Similarity :", S1)

S2 = wv.similarity(w1 = "Bread", w2 = "Wheat")
print("\nSimilarity :", S2)


# Most Similar Words

S3 = wv.most_similar("Good")
print("\nMost Similar Words :", S3)

S4 = wv.most_similar("Worst")
print("\nMost Similar Words :", S4)

S5 = wv.most_similar(positive = ["King","Woman"], negative = ["Man"])
print("\nSimilar Words :", S5)

S6 = wv.most_similar(positive = ["France","Berlin"], negative = ["Paris"])
print("\nSimilar Words :", S6)


# Doesn't Match

S7 = wv.doesnt_match(["facebook", "cat", "google", "microsoft"])
print("\nDoesn't Match :", S7)

S8 = wv.doesnt_match(["Cricket", "Football", "Tennis", "Hockey"])
print("\nDoesn't Match :", S8)
```

```python

import pandas as pd
import spacy



# Define the Dataset

df = pd.read_csv("Fake_Real_Data.csv")
print(df.head())
print("\nShape :",df.shape)


# Count of each category

Count = df["label"].value_counts()
print("\nCount :", Count)


# Convert Text into Numbers in LabelColumn

Target = {"Fake":0, "Real":1}
df["New_label"] = df["label"].map(Target)
print("\n",df.head())


# Preprocess the text to remove stop words, punctuations and get lemma for each word

nlp = spacy.load("en_core_web_lg")

def preprocess(text):

  doc = nlp(text)
  filtered_tokens = []
  for token in doc:
    if token.is_stop or token.is_punct:
      continue
    filtered_tokens.append(token.lemma_)

  return wv.get_mean_vector(filtered_tokens)


# Create a New column

df["vector"] = df["Text"].apply(lambda text: preprocess(text))
print("\n",df.head())
```

```python

from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt



# Split the dataset to Train & Test data

X_train,X_test,y_train,y_test = train_test_split(df["vector"].values, df["New_label"], test_size = 0.2, random_state = 2022, stratify = df.New_label)

print("Shape of X_train before reshaping: ", X_train.shape)
print("Shape of X_test before reshaping: ", X_test.shape)


# Convert to 2D array

X_train_2D = np.stack(X_train)
X_test_2D = np.stack(X_test)

print("Shape of X_train after reshaping: ", X_train_2D.shape)
print("Shape of X_test after reshaping: ", X_test_2D.shape)


# Train using Gradient Boosting Classifier

GBC = GradientBoostingClassifier()
GBC.fit(X_train_2D, y_train)
y_pred = GBC.predict(X_test_2D)


# Classification report

Report = classification_report(y_test, y_pred)
print("\nClassification Report :\n", Report)


# Confusion Matrix

Matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix :\n", Matrix)
print("\n")


# Plot the Confusion Matrix

plt.figure(figsize = (10,7))
sns.heatmap(Matrix, annot = True, fmt = "d")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
